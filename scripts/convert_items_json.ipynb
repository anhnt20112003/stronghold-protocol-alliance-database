```python
# Convert items.json to a JSON object with a single `shopiteminfo` list
This notebook reads the raw `src/data/items.json`, parses it into a Python dictionary, converts the values into a list, and writes a new JSON file containing a single `shopiteminfo` array.
</VSCode.Cell>
<VSCode.Cell id="2" language="python">
from pathlib import Path

source_path = Path(r"c:\Users\Admin\Desktop\spa\src\data\items.json")
raw_text = source_path.read_text(encoding="utf-8")
raw_text[:400]
</VSCode.Cell>
<VSCode.Cell id="3" language="markdown">
## Parse item entries into Python dictionary
The source file contains top-level object entries without an enclosing object. This cell normalizes that text into valid JSON and parses it.
</VSCode.Cell>
<VSCode.Cell id="4" language="python">
import re
import json

# Normalize and wrap the raw text into a valid JSON object.
# Remove any trailing commas and preserve internal object formatting.

normalized = raw_text.strip()
if not normalized.startswith("{"):
    normalized = "{" + normalized
if not normalized.endswith("}"):
    normalized = normalized + "}"

# Replace any single trailing comma before final closing brace.
normalized = re.sub(r",\s*}\s*$", "}", normalized, flags=re.DOTALL)

items_dict = json.loads(normalized)
len(items_dict)
</VSCode.Cell>
<VSCode.Cell id="5" language="markdown">
## Create `shopiteminfo` list
Convert the parsed dictionary values into a list and wrap them inside a new object with the `shopiteminfo` key.
</VSCode.Cell>
<VSCode.Cell id="6" language="python">
output = {"shopiteminfo": list(items_dict.values())}
len(output["shopiteminfo"])
</VSCode.Cell>
<VSCode.Cell id="7" language="markdown">
## Write transformed JSON file
Save the transformed object to a new file that can be used in the project.
</VSCode.Cell>
<VSCode.Cell id="8" language="python">
output_path = Path(r"c:\Users\Admin\Desktop\spa\src\data\items_transformed.json")
output_path.write_text(json.dumps(output, indent=2, ensure_ascii=False), encoding="utf-8")
output_path
```
